```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/Market_Basket_Optimisation.csv")
df.head()
```

```python
transactions = []
for i in range(0,len(df)):
  transactions.append([str(df.values[i,j])for j in range(0,20) if str(df.values[i,j])!='0'])
```

```python
!pip install apyori
```

```python
from apyori import apriori
rules = apriori(transactions = transactions,min_support = 0.003,min_confidence = 0.2,min_lift = 3,min_length = 2,max_length = 2)
rules = list(rules)
```

```python
results = []

for rule in rules:
    for stat in rule.ordered_statistics:
        if stat.items_base and stat.items_add:
            lhs = list(stat.items_base)[0]   # extract string
            rhs = list(stat.items_add)[0]    # extract string

            results.append([
                lhs,
                rhs,
                rule.support,
                stat.confidence,
                stat.lift
            ])

import pandas as pd
result = pd.DataFrame(
    results,
    columns=["Left Product", "Right Product", "Support", "Confidence", "Lift"]
)

result
```
